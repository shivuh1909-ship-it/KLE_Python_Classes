```python
#multilevel inheritance

class grand_father:
  def irayya(self):
    print("I'm 89 years old")

class father(grand_father):
  def basayya(self):
    print("I'm 50 years old")

class children(father):
  def shiva(self):
    print("I'm 18 years old")

obj = children()
obj.irayya()
obj.basayya()
obj.shiva()
```

```python
class dictionary:
  def __init__(self): #correct constructor
  #Stored words in a dictionary(key =word,valu=mening)
    self.words = {}

  def add_word(self, word, meaning):
    #prevent duplicate keys
   if word in self.words:
      print(f"'{word}' already exists in dictionary. Duplicate not allowed!")
   else:
      self.words[word] = meaning
      print(f"'{word}' added successfully!")

  def search_word(self, word):
    #search for a word and display result in table format
    if word in self.words:
      print("\n===== Search Result =====")
      print(f"{'Word':<15}|{'Meaning'}")
      print("-" * 40)
      print(f"{word:<15} | {self.words[word]}")
      print("-" * 40)
    else:
      print(f"'{word}' not found in dictionary.")

  def show_all(self):
    #show all words alphabetically
    if not self.words:
      print("Dictionary is empty.")
    else:
      print("\n===== My Dictionary (alphabetical order)=====")
      print(f"{'Word':<15} | {'Meaning'}")
      print("-" * 40)
      for word in sorted(self.words.keys()):
        print(f"{word:<15} | {self.words[word]}")
      print("-" * 40)


  def update_word(self, word, new_meaning):
    #update word if it exists
    if word in self.words:
      self.words[word] = new_meaning
      print(f"'{word}' updated successfully!")
    else:
      print(f"'{word}' not found. Cannot update.")

def main():
  my_dict = dictionary()

  while True:
    print("\n====== Digital Dictionary ======")
    print("1. Add Word")
    print("2. Search Word")
    print("3. Show All Words")
    print("4. Update Word")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
      word = input("Enter word: ")
      meaning = input("Enter meaning: ")
      my_dict.add_word(word, meaning)

    elif choice == "2":
      word = input("Enter word to search: ")
      my_dict.search_word(word)

    elif choice == "3":
      my_dict.show_all()

    elif choice == "4":
      word = input("Enter word to update: ")
      new_meaning = input("Enter new meaning: ")
      my_dict.update_word(word, new_meaning)

    elif choice == "5":
      print("Exiting Digital Dictionary... Goodbye!")
      break

    else:
      print("Invalid choice. Please try again.")

#Run the program
if __name__ == "__main__":
 main()
```
